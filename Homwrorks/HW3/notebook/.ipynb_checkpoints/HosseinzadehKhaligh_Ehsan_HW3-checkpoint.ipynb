#### i. Research

Time-series classification involves the analysis of time-series data to identify patterns or classify the data into different categories. Time-domain features are one of the most commonly used types of features for time-series classification. Here are some examples of commonly used time-domain features:

Mean: The average value of the time series.

Standard deviation: The measure of the dispersion of the data.

Skewness: A measure of the asymmetry of the distribution of the data.

Kurtosis: A measure of the peakiness of the distribution of the data.

Maximum: The highest value in the time series.

Minimum: The lowest value in the time series.

Range: The difference between the maximum and minimum values.

Variance: The average of the squared differences from the mean.

Autocorrelation: A measure of the similarity between a time series and a lagged version of itself.

Cross-correlation: A measure of the similarity between two time series at different lags.

Entropy: A measure of the unpredictability or randomness of the data.

Energy: The sum of the squares of the data values.

Root mean square (RMS): The square root of the average of the squared values.

Zero crossing rate: The number of times the signal crosses the zero axis.

Time-domain fractal dimension: A measure of the complexity of the time series.

These time-domain features can be used in combination with other types of features, such as frequency-domain or wavelet-based features, to improve the accuracy of time-series classification algorithms.

#### iii. Standard Deviation

```python
#we need to first merge two dfs 
train_test_df = df_main_train.append(df_main_test, ignore_index=True)
train_test_df.head()
```

```python
df_iii = pd.DataFrame(columns = ['feature', 'std', 'interval'])

for col_name in list(train_test_df.columns.values):
    
    #print(col_name)
    
    feature_val_list = train_test_df[col_name].to_numpy()
    feature_val_list = [np.float64(x) for x in feature_val_list]
    feature_val_list = np.asarray(feature_val_list)
    
    #debug
    #print(feature_val_list)
    #print(type(feature_val_list))
    #print(type(feature_val_list[0]))
    
    bootstrap_std_devs = bs.bootstrap(feature_val_list, stat_func=bs_stats.std, alpha=0.1, num_iterations=10000)
    
    #print(bootstrap_std_devs, '\n') #prints std (lower_bound, upper_bound)
    
    #print(bootstrap_std_devs.value)
    #print(bootstrap_std_devs.lower_bound)
    #print(bootstrap_std_devs.upper_bound)
    
    df_iii = df_iii.append({'feature' : col_name, 'std' : bootstrap_std_devs.value, 'interval' : str(round(bootstrap_std_devs.lower_bound, 4))+','+str(round(bootstrap_std_devs.upper_bound, 4))}, ignore_index = True)

display(df_iii)
```
